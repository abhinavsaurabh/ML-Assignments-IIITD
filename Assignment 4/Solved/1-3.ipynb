```python
feature_cols = ['sepal length', 'sepal width', 'petal length', 'petal width']
target = ['class']
X = df[feature_cols]
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y = le.fit_transform(df['class'])
```
